```python
import pandas as pd
import numpy as np
import re
import string
import torch
from torch import nn
from sklearn.model_selection import train_test_split
from datasets import Dataset, DatasetDict
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
import evaluate
import matplotlib.pyplot as plt
import seaborn as sns
import nltk

# ==============================================
# 1. LOAD DATASET
# ==============================================
df = pd.read_csv("data.csv")  # Pastikan file ada
print("Jumlah data:", len(df))
print(df.head())

text_col = "comment"
rating_col = "rating"

# ==============================================
# 2. LABELING SENTIMEN BERDASARKAN RATING
# ==============================================
def rating_to_label(rating):
    if rating <= 2:
        return "negatif"
    elif rating == 3:
        return "netral"
    else:
        return "positif"

df["label"] = df[rating_col].apply(rating_to_label)

# Label encoding
label2id = {"negatif": 0, "netral": 1, "positif": 2}
id2label = {v: k for k, v in label2id.items()}
df["label_id"] = df["label"].map(label2id)

print("\nDistribusi label:")
print(df["label"].value_counts())

# ==============================================
# 3. PREPROCESSING TEKS
# ==============================================
nltk.download("stopwords")
from nltk.corpus import stopwords
stop_words = set(stopwords.words("indonesian"))

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+|#\w+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text.strip()

df["clean_comment"] = df[text_col].apply(clean_text)
print("\nContoh hasil preprocessing:")
print(df[["comment", "clean_comment"]].head())

# ==============================================
# 4. SPLIT DATASET (80:10:10)
# ==============================================
train_df, temp_df = train_test_split(df, test_size=0.2, stratify=df["label_id"], random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df["label_id"], random_state=42)
print(f"\nTrain: {len(train_df)}, Validation: {len(val_df)}, Test: {len(test_df)}")

# Konversi ke Dataset Hugging Face
train_dataset = Dataset.from_pandas(train_df[["clean_comment", "label_id"]])
val_dataset = Dataset.from_pandas(val_df[["clean_comment", "label_id"]])
test_dataset = Dataset.from_pandas(test_df[["clean_comment", "label_id"]])

dataset = DatasetDict({
    "train": train_dataset,
    "validation": val_dataset,
    "test": test_dataset
})

# ==============================================
# 5. TOKENISASI DENGAN INDOBERT
# ==============================================
model_name = "indobenchmark/indobert-base-p1"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_fn(examples):
    return tokenizer(
        examples["clean_comment"], 
        padding="max_length", 
        truncation=True, 
        max_length=128
    )

tokenized_datasets = dataset.map(tokenize_fn, batched=True)

# Setelah tokenisasi
tokenized_datasets = tokenized_datasets.rename_column("label_id", "labels")

# Hapus kolom yang tidak diperlukan
for col in ["clean_comment", "__index_level_0__", "username", "rating"]:
    if col in tokenized_datasets["train"].column_names:
        tokenized_datasets = tokenized_datasets.remove_columns([col])

# Set format ke torch
tokenized_datasets.set_format("torch")


# ==============================================
# 6. HITUNG CLASS WEIGHTS UNTUK LOSS
# ==============================================
from sklearn.utils.class_weight import compute_class_weight

class_weights_np = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(df["label_id"]),
    y=df["label_id"]
)
class_weights = torch.tensor(class_weights_np, dtype=torch.float)
print("Class weights:", class_weights)

# ==============================================
# 7. LOAD MODEL INDO BERT
# ==============================================
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=3,
    id2label=id2label,
    label2id=label2id
)

# ==============================================
# 8. METRIK EVALUASI
# ==============================================
metric_acc = evaluate.load("accuracy")
metric_f1 = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    acc = metric_acc.compute(predictions=preds, references=labels)
    f1 = metric_f1.compute(predictions=preds, references=labels, average="weighted")
    return {"accuracy": acc["accuracy"], "f1": f1["f1"]}

# ==============================================
# 9. CUSTOM TRAINER DENGAN WEIGHTED LOSS
# ==============================================
class WeightedTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.logits
        loss_fct = nn.CrossEntropyLoss(weight=class_weights.to(logits.device))
        loss = loss_fct(logits.view(-1, model.config.num_labels), labels.view(-1))
        return (loss, outputs) if return_outputs else loss


# ==============================================
# 10. TRAINING ARGUMENTS
# ==============================================
training_args = TrainingArguments(
    output_dir="./results_all",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    remove_unused_columns=False
)

trainer = WeightedTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

# ==============================================
# 11. TRAINING
# ==============================================
trainer.train()

# ==============================================
# 12. EVALUASI AKHIR
# ==============================================
preds_output = trainer.predict(tokenized_datasets["test"])
preds = np.argmax(preds_output.predictions, axis=-1)

from sklearn.metrics import classification_report, confusion_matrix
print("\n=== HASIL EVALUASI PADA TEST SET ===")
print(classification_report(test_df["label_id"], preds, target_names=id2label.values()))

# Visualisasi Confusion Matrix
plt.figure(figsize=(6,4))
sns.heatmap(
    confusion_matrix(test_df["label_id"], preds),
    annot=True, fmt="d", cmap="Blues",
    xticklabels=id2label.values(), yticklabels=id2label.values()
)
plt.title("Confusion Matrix - Test Data")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ==============================================
# 13. SIMPAN MODEL & TOKENIZER
# ==============================================
trainer.save_model("./indoBERT_sentiment_shopee_all")
tokenizer.save_pretrained("./indoBERT_sentiment_shopee_all")
print("✅ Model IndoBERT Sentiment Shopee berhasil disimpan!")
```
